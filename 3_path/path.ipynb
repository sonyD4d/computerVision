# Computer vision : Path application

        - Calculate the distance between indian airports by using path 
        - Path calculated using Euclidean distance
        - Assuming 4,d and 8 connected

```python
# Libs
import cv2
import matplotlib.pyplot as plt
import math
```

```python
count = 0      # to keep track of point A and B
cords = []     # to keep track of coordinates A and B
```

```python
# Euclidean distance
def e_dist(cords):
    return ( (cords[0][0] - cords[1][0])**2 + (cords[0][1] - cords[1][1])**2 )**0.5
```

```python
# On click event handler for mouse clicks
def onClickEvent(event, x, y, flags, params):
    global count
    if event == cv2.EVENT_LBUTTONDOWN:
        cords.append((x, y))
        count+=1
        font = cv2.FONT_HERSHEY_SIMPLEX
        cv2.putText(img, str(x) + ',' +str(y), (x,y), font,1, (255, 0, 0), 2)
        cv2.imshow('Pick point', img)
        if(count == 2):
            cv2.destroyAllWindows()
```

```python
# main 
if __name__=="__main__":
    global count
    img = cv2.imread('2_image.jpg', 1)
    font = cv2.FONT_HERSHEY_SIMPLEX
    cv2.imshow('Pick point', img)
    cv2.setMouseCallback('Pick point', onClickEvent)
    cv2.waitKey(0)
    print("Euclidean distance\t: " + str(math.ceil(e_dist(cords))) + " Pixels")
    print("Distance\t\t: " + str(math.ceil(e_dist(cords)*3.91)) + " KM")
    final = cv2.imread('2_image.jpg',1)
    final = cv2.cvtColor(final, cv2.COLOR_BGR2RGB)
    cv2.arrowedLine(final, cords[0], cords[1], (0, 0, 255), thickness=5, tipLength = 0.06) 
    cv2.putText(final, '  ( ' + str(cords[0][0]) + ',' +str(cords[0][1]) + ' )', cords[0] , font,1, (0, 0, 255), 2)
    cv2.putText(final, '  ( ' + str(cords[1][0]) + ',' +str(cords[1][1]) + ' )', cords[1] , font,1, (0, 0, 255), 2)
    plt.figure(figsize=(20,11))
    plt.axis("off")
    plt.imshow(final)
    plt.show()
    count = 0
    cords.clear()
```
